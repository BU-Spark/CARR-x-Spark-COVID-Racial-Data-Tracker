```python
NY_df = pd.read_csv('https://docs.google.com/spreadsheets/d/1xF1cdazAejukTt2M6-aAXWG6fivTHFqt8rurF8euXOE/export?format=csv&gid=43720681', header=0)
#imports public csv directly into program as pandas dataframe
# NY_df = pd.read_csv('/Users/hoperuse/Downloads/Race Data Entry - CRDT.csv')
NY_df.head(10)
```

```python
def convert_date(val):
    # function to convert integer values for Date on public sheet to date strings
    year = val // 10000
    month = (val % 10000)//100
    day = (val%10000)%100
    if month < 10:
        month = '0'+str(month)
    if day < 10:
        day = '0' + str(day)
    date = str(month) + '/' + str(day) + '/'+ str(year)
    return date
```

```python
NY_df["Date"] = NY_df['Date'].apply(convert_date)
# NY_df['Date'] = pd.to_datetime(NY_df['Date'], errors='ignore')
# applies date conversion function to date column on public csv dataframe

NY_df.head(10)
```

```python
# NY_df.dtypes
# list the types of each column on dataframe -- note that numerical values like case total are listed as objects 
# instead of numeric values
```

```python
NY_df = NY_df.sort_values(by='Date')
# sorts dataframe into ascending order by date, helps visualize data chronologically later
NY_df.head(10)
```

```python
NY_df = pd.DataFrame(NY_df[NY_df['State'] == 'NY'])
# sort and filter dataframe to only rows for NY

NY_df.tail(10)
```

```python
# Correcting column dtypes from 'object' to floats
NY_df['Cases_Total'] = pd.to_numeric(NY_df['Cases_Total'], errors='coerce')

NY_df['Cases_White'] = pd.to_numeric(NY_df['Cases_White'], errors='coerce')

NY_df['Cases_Ethnicity_Unknown'] = pd.to_numeric(NY_df['Cases_Ethnicity_Unknown'], errors='coerce')

NY_df.dtypes
```

```python
totalCases = NY_df['Deaths_Total'] + NY_df['Cases_Total']

# pctChangeDW_ofTotal = NY_df['Deaths_White'] / NY_df['Deaths_Total']
# pctChangeDW_ofTotalShift = (NY_df['Deaths_White'].shift(-1) / NY_df['Deaths_Total'].shift(-1))
NY_df['Deaths White % Total'] = ((NY_df['Deaths_White'] / NY_df['Deaths_White'] + totalCases)*100).diff()

# pctChangeDB_ofTotal = NY_df['Deaths_Black'] / NY_df['Deaths_Total']
# pctChangeDB_ofTotalShift = (NY_df['Deaths_Black'].shift(-1) / NY_df['Deaths_Total'].shift(-1))
NY_df['Deaths Black % Total'] = ((NY_df['Deaths_Black'] / (NY_df['Deaths_Black'] + totalCases))*100).diff()

# pctChangeDO_ofTotal = NY_df['Deaths_Other'] / NY_df['Deaths_Total']
# pctChangeDO_ofTotalShift = (NY_df['Deaths_Other'].shift(-1) / NY_df['Deaths_Total'].shift(-1))
NY_df['Deaths Other % Total'] = ((NY_df['Deaths_Other'] / (NY_df['Deaths_Other'] + totalCases))*100).diff()

# pctChangeDU_ofTotal = NY_df['Deaths_Unknown'] / NY_df['Deaths_Total']
# pctChangeDU_ofTotalShift = (NY_df['Deaths_Unknown'].shift(-1) / NY_df['Deaths_Total'].shift(-1))
NY_df['Deaths Unknown % Total'] = ((NY_df['Deaths_Unknown'] / (NY_df['Deaths_Unknown'] + totalCases))*100).diff()

# NY_df.head()
# NY_df['Deaths White % Total']
# NY_df['Deaths White % total change'] = 
```

```python
NY_df['Cases_Total'] = NY_df['Cases_Total'].diff()

NY_df['Cases_Unknown'] = NY_df['Cases_Unknown'].diff()

NY_df['Cases_Ethnicity_Unknown'] = NY_df['Cases_Ethnicity_Unknown'].diff()

NY_df['Deaths_Total'] = NY_df['Deaths_Total'].diff()

NY_df['Deaths_White'] = NY_df['Deaths_White'].diff()

NY_df['Deaths_Black'] = NY_df['Deaths_Black'].diff()

NY_df['Deaths_LatinX'] = NY_df['Deaths_LatinX'].diff()

NY_df['Deaths_Asian'] = NY_df['Deaths_Asian'].diff()

NY_df['Deaths_Other'] = NY_df['Deaths_Other'].diff()

NY_df['Deaths_Unknown'] = NY_df['Deaths_Unknown'].diff()

NY_df['Deaths_Ethnicity_Hispanic'] = NY_df['Deaths_Ethnicity_Hispanic'].diff()

NY_df['Deaths_Ethnicity_NonHispanic'] = NY_df['Deaths_Ethnicity_NonHispanic'].diff()

NY_df['Deaths_Ethnicity_Unknown'] = NY_df['Deaths_Ethnicity_Unknown'].diff()
NY_df.tail(10)
```

```python
# percentage changes of self of each column overtime
NY_df['% Change Total Cases'] = (NY_df['Cases_Total'].pct_change() * 100)

NY_df['% Change Cases Unknown'] = (NY_df['Cases_Unknown'].pct_change() * 100)

NY_df['% Change Cases Ethnicity Unknown'] = (NY_df['Cases_Ethnicity_Unknown'].pct_change() * 100)

NY_df['% Change Total Deaths'] = (NY_df['Deaths_Total'].pct_change() * 100)

NY_df['% Change Total Deaths (White)'] = (NY_df['Deaths_White'].pct_change() * 100)

NY_df['% Change Total Deaths (Black)'] = (NY_df['Deaths_Black'].pct_change() * 100)

NY_df['% Change Total Deaths (LatinX)'] = (NY_df['Deaths_LatinX'].pct_change() * 100)

NY_df['% Change Total Deaths (Asian)'] = (NY_df['Deaths_Asian'].pct_change() * 100)

NY_df['% Change Total Deaths (Other)'] = (NY_df['Deaths_Other'].pct_change() * 100)

NY_df['% Change Total Deaths (Unknown)'] = (NY_df['Deaths_Unknown'].pct_change() * 100)

NY_df['% Change Total Deaths by Ethnicity Hispanic'] = (NY_df['Deaths_Ethnicity_Hispanic'].pct_change() * 100)

NY_df['% Change Total Deaths by Ethnicity NonHispanic'] = (NY_df['Deaths_Ethnicity_NonHispanic'].pct_change() * 100)

NY_df['% Change Total Deaths by Ethnicity Unknown'] = (NY_df['Deaths_Ethnicity_Unknown'].pct_change() * 100)
```
